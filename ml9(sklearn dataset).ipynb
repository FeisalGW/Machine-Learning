# Californian House

```python
import pandas as pd 
import numpy as np
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
```

```python
dataCH = fetch_california_housing()
dir(dataCH)
```

```python
dataCH.data[0]
```

```python
dataCH['feature_names']
```

```python
dfCH = pd.DataFrame(
    dataCH.data,
    columns = dataCH['feature_names']
)
dfCH['PRICE'] = dataCH.target
dfCH
```

```python
dfCH.corr()
```

```python
plt.imshow(dfCH.corr())
plt.colorbar
```

## Spliting Data Sets
### 90% Training = 10% Testing

```python
x = dfCH[['MedInc','HouseAge','AveRooms','AveBedrms',
       'Population','AveOccup','Latitude','Longitude']]
y = dfCH['PRICE']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .1)
```

```python
x_test.iloc[0]
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#prediksi
model.predict([x_test.iloc[0]])
```

```python
# score accuracy
model.score(x_train, y_train)
```

## Save Model Machine Learning:
### 1. pickle

```python
import pickle
with open('modelPickle.pkl', 'wb') as modelku:
    pickle.dump(model, modelku)
```

### 2. joblib

```python
import joblib
joblib.dump(model, 'modeljoblib.joblib')
```
