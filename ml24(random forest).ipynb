 # Regression with Random Forest vs Decision Tree

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

### 1. creat data

```python
x = np.sort(np.random.randn(2000))
y = np.cos(x)
```

### 2. create model

```python
modelDT = DecisionTreeRegressor(max_depth = 5)
modelRF = RandomForestRegressor(n_estimators = 5)

modelDT.fit(x.reshape(-1, 1), y)
modelRF.fit(x.reshape(-1, 1), y)
```

### 3. find best y from decission tree model & random forest model

```python
yDT = modelDT.predict(x.reshape(-1,1))
yRF = modelRF.predict(x.reshape(-1,1))
```

### 4. plotting & comparing result

```python
plt.plot(x, y, 'y.')
plt.plot(x, yDT, 'g-')
plt.plot(x, yRF, 'r-')
plt
```
