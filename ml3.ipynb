```python
import numpy as np 
import pandas as pd 
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt  
```

```python
##
```

```python
dataBoston = load_boston()
df = pd.DataFrame(
    dataBoston['data'],
    columns = dataBoston ['feature_names']
)
df['price'] = dataBoston['target']
df 
```

```python
korelasi = df.corr()
korelasi
```

```python
korelasi = df.corr()
plt.imshow(korelasi, cmap='BuPu_r')
plt.colorbar()
plt.xticks(np.arange(14), list(df.columns))
plt.yticks(np.arange(14), list(df.columns))
plt.show
```

```python
model = LinearRegression()
model.fit(df[['ZN','CHAS','RM','DIS','B']], df['price'])
```

```python
model = LinearRegression()
model.fit(df[['ZN','CHAS','RM','DIS','B']], df['price'])
model.predict(df[['ZN','CHAS','RM','DIS','B']])
```
