```python
# machine learning fro classification
import numpy as np 
import pandas as pd 
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt 
```

```python
dataIris = load_iris()
print(dir(dataIris))
```

```python
print(dataIris['feature_names'])
```

```python
dfIris = pd.DataFrame(
    dataIris['data'],
    columns = dataIris['feature_names']
)
dfIris
```

```python
dfIris = pd.DataFrame(
    dataIris['data'],
    columns = ['SL','SW','PL','PW']
)
dfIris['target'] = dataIris['target']
dfIris['spesies'] = dfIris['target'].apply(
    lambda x: dataIris['target_names'][x]
)
dfIris
```

```python
dfSetosa = dfIris[dfIris['target'] == 0]
dfVersicolor = dfIris[dfIris['target'] == 1]
dfVirginica = dfIris[dfIris['target'] == 2]
```

```python
fig = plt.figure('Data Iris', figsize = (6,6))

# plot sepal length vs sepal width
plt.subplot(121)
plt.scatter(dfSetosa['SL'], dfSetosa['SW'], marker='o', color='r')
plt.scatter(dfVersicolor['SL'], dfVersicolor['SW'], marker='o', color='g')
plt.scatter(dfVirginica['SL'], dfVirginica['SW'], marker='o', color='b')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(['Setosa','Versicolor','Virginica'])
plt.grid(True)

# plot petal length vs petal width
plt.subplot(122)
plt.scatter(dfSetosa['SL'], dfSetosa['PW'], marker='o', color='r')
plt.scatter(dfVersicolor['SL'], dfVersicolor['PW'], marker='o', color='g')
plt.scatter(dfVirginica['SL'], dfVirginica['PW'], marker='o', color='b')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(['Setosa','Versicolor','Virginica'])
plt.grid(True)
```
