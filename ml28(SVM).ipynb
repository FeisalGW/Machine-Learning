# Support Vektor Machines

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

### 1. create data 

```python
dataIris = load_iris()
df = pd.DataFrame(
    dataIris['data'],
    columns = ['SL', 'SW', 'PL','PW']
)
df['target'] = dataIris['target']
df['spesies'] = df['target'].apply(
    lambda row: dataIris['target_names'][row]
)
```

```python
df.head(3)
```

### 2. create model SVM 

```python
from sklearn.svm import SVC
model = SVC(gamma = 'auto', probability = True)
model.fit(df[['SL', 'SW', 'PL','PW']], df['target'])
model.score(df[['SL', 'SW', 'PL','PW']], df['target'])
df['pred'] = model.predict(df[['SL', 'SW', 'PL','PW']])
df
```

### 3. predict

```python
model.predict([[2, 1.2, 3, 3.2]])
```

```python
model.predict_proba([[2, 1.2, 3, 3.2]])
```

### 4. plotting

```python
modelS = SVC(gamma = 'auto', probability = True)
modelP = SVC(gamma = 'auto', probability = True)
sepal = dataIris['data'][:, :2]
sl, sw = sepal[:, 0], sepal[:, 1]
petal = dataIris['data'][:, 2:]
pl, pw = petal[:, 0], petal[:, 1]

modelS.fit(sepal, dataIris['target'])
modelP.fit(petal, dataIris['target'])
```

```python
#mgrid = meshgrid
def mgrid(x, y):
    x_min = x.min() - 1
    x_max = x.max() + 1
    y_min = y.min() - 1
    y_max = y.max() + 1
    xx, yy = np.meshgrid(
        np.arange(x_min, x_max, .02),
        np.arange(y_min, y_max, .02)
    )
    return xx, yy

sl, sw = mgrid(sl, sw)
pl, pw = mgrid(pl, pw)

#function plotting
def svmplot(axisfig, model, length, width):
    p = model.predict(np.c_[length.ravel(), width.ravel()])
    p = p.reshape(length.shape)
    hasil = axisfig.contourf(length, width, p)
    return hasil
```

```python
fig = plt.figure

ax = plt.subplot(121)
svmplot(ax, modelS, sl, sw)
plt.plot(
    df[df['target'] == 0]['SL'],
    df[df['target'] == 0]['SW'],
    'ro'
)
plt.plot(
    df[df['target'] == 1]['SL'],
    df[df['target'] == 1]['SW'],
    'bo'
)
plt.plot(
    df[df['target'] == 2]['SL'],
    df[df['target'] == 2]['SW'],
    'yo'
)

ax = plt.subplot(122)
svmplot(ax, modelP, pl, pw)
plt.plot(
    df[df['target'] == 0]['PL'],
    df[df['target'] == 0]['PW'],
    'ro'
)
plt.plot(
    df[df['target'] == 1]['PL'],
    df[df['target'] == 1]['PW'],
    'bo'
)
plt.plot(
    df[df['target'] == 2]['PL'],
    df[df['target'] == 2]['PW'],
    'yo'
)

plt.show()
```
