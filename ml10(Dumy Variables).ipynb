## Dummy Variables

```python
import pandas as pd
import numpy as np
```

### 1. read dataset (data.csv)

```python
data = pd.read_csv('ml10.csv')
data
```

### 2. create dummy var for data['kota']

```python
dataDum = pd.get_dummies(data['kota'])
dataDum
```

### 3. create final dataframe

```python
dataFinal = pd.concat([data, dataDum], axis='columns')
dataFinal
```

```python
dataFinal = dataFinal.drop(['kota'], axis=1)
dataFinal
```

### 4. linear regression

```python
y = dataFinal['harga']
x = dataFinal.drop(['harga'], axis='columns')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)
```

```python
model.score(x, y)*100
```

### 5. prediction

```python
#Harga jika luas 1000 di Depok
model.predict([[1000, 0, 1, 0]])[0]
```
